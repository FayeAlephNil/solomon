```python
import solomon.examples as ex
import solomon.modular as smod
def primes_method5(n):
    out = list()
    sieve = [True] * (n+1)
    for p in range(2, n+1):
        if (sieve[p]):
            out.append(p)
            for i in range(p, n+1, p):
                sieve[i] = False
    return out
```

```python
(S,rep,orb) = ex.alternating_genus_1(5)
primes = primes_method5(5)
mod_reps_orbs = smod.mod_rep_orbs(rep,orb, lst=primes)
mod_orbs = [(primes[i],orb) for (i,(_,orb)) in enumerate(mod_reps_orbs)]
```

```python
mod_orbs[0][1].advance_until(N=200)
len(mod_orbs[0][1].digraph.nodes)
```

```python
advanced_orbs = smod.advance_all_orbs([orb for (_, orb) in mod_orbs])
list(advanced_orbs[0].digraph.nodes)[12].underlying_lst
```
