```python
E = EisensteinIntegers()
omega = E.gens()[1]
F = E.residue_field(2)
omegabar = F.gens()[0]
def burau_mod_2_cube_root(b):
    return b.burau_matrix(reduced=True).subs(t=omegabar)
```

```python
def all_vecs(n):
    V = VectorSpace(F,n-1)
    return V
def get_orb_with_braids(n,braids):
    orb = [burau_mod_2_cube_root(b) * e1 for b in braids]
    return orb

def construct_orbit_mod_2_cube_root(m,start=None,breadth=False,N=100000):
    BGroup = BraidGroup(m, 'σ')
    gens = BGroup.gens()
    Gr = DiGraph(loops=True,multiedges=True)
    if not(start):
        start = all_vecs(m).basis()[0]

    seen = [start]
    new_vtxs = [(start,0)]
    Gr.add_vertex()
    num_loops = 0
    i = 1
    completed_verts = []
    while new_vtxs and (i < N):
        if breadth:
            (v,k) = new_vtxs.pop(0)
        else:
            (v,k) = new_vtxs.pop()
        for g in gens:
            w = burau_mod_2_cube_root(g)* v
            if not (w in seen):
                seen.append(w)
                new_vtxs.append((w,i))
                j = i
                Gr.add_vertex()
                i += 1
            else:
                j = seen.index(w)
            Gr.add_edge(k,j, label=str(g))
        completed_verts.append(k)
        num_loops += 1
    return (Gr,seen)
```

```python
n=7
PF = ProjectiveSpace(n-2,F)
e = list(PF)[0]
print(e)
print(construct_orbit_mod_2_cube_root(n,start=e))
print(len(list(PF)))
```
